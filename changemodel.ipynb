```python
student_purchases.value_counts('purchase_type')
```

```python
sale_duration = {'Winder Sale 2022':[datetime(2022,1,17), datetime(2022,1,28)],
                'Spring Special Offer 2022':[datetime(2022,3,21), datetime(2022,3,31)],
                'Data Science Summer Campaign 2022':[datetime(2022,5,20), datetime(2022,6,1)],
                'Most Wanted Campaign':[datetime(2022,7,18), datetime(2022,7,29)],
                'Platform gamification':[datetime(2022,9,16), datetime(2022,9,16)],
                'Data Science Hero Campaign':[datetime(2022,9,19), datetime(2022,9,30)]}
```

```python
sale_purchase = {}
for t,(s,e) in sale_duration.items():
    sale_student = student_purchases[(student_purchases['date_purchased']>=s)&(student_purchases['date_purchased']<=e)]
    sale_purchase[t] = sale_student

sum_d = 0
for t, d in sale_purchase.items():
    print(t, len(d))
    sum_d += len(d)
    
print(sum_d)
```

```python
sum_d / len(student_purchases)
```
